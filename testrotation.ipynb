```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("5.jpg")

rotated1 =cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotated1)
plt.show()


# Get image center
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

# Rotation matrix: angle=45, scale=1.0
M = cv2.getRotationMatrix2D(center, 180, 1.0)

# Rotate the image
rotated = cv2.warpAffine(img, M, (w, h))

plt.imshow(rotated)
```
