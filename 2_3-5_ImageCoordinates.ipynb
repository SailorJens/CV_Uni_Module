```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
apple_image = cv2.imread('apple.jpg')
apple_image = cv2.cvtColor(apple_image, cv2.COLOR_BGR2RGB)
```

```python
height, width, channels = apple_image.shape
print(f'Image dimensions: {width}x{height}, Channels: {channels}')
```

```python
# Get a pixel value at (50, 100)
red, green, blue = apple_image[100, 50]
print(f'Pixel at (50, 100) - R: {red}, G: {green}, B: {blue}')
```

```python
apple_image = cv2.imread('apple.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
height, width = apple_image.shape
print(f'Image dimensions: {width}x{height}')
```

```python
# Get a pixel value at (50, 100)
pixel_value = apple_image[100, 50]
print(f'Pixel value at (50, 100): {pixel_value}')
```

```python
plt.imshow(apple_image, cmap='gray')
plt.title('Grayscale Apple Image')
```

```python
# Region of interest 0|0, 100|100
region = apple_image[0:100, 0:100]
plt.figure()
plt.imshow(region, cmap='gray')
plt.title('Region of Interest (0:100, 0:100)')
```

```python
# Create a copy of the image and modify a region
new_image = apple_image.copy()
```

```python
new_image[40:100, 150:300] = 255  # Set a square region to white
plt.figure()
plt.imshow(new_image, cmap='gray')
plt.title('Modified Image with White Square Region')
```

```python
# Make another copy and modify a different region
another_image = apple_image.copy()
darker_region = another_image[40:100, 150:300] * 0.6  # Dim a square region
another_image[40:100, 150:300] = darker_region.astype(int)
plt.figure()
plt.imshow(another_image, cmap='gray')
plt.title('Modified Image with Dimmed Square Region')
```

```python
# Doing the same with the original color image
color_image = cv2.imread('apple.jpg')
color_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)
color_image_copy = color_image.copy()
darker_region = color_image_copy[40:100, 150:300] * 0.6
color_image_copy[40:100, 150:300] = darker_region.astype(int)
plt.figure()
plt.imshow(color_image_copy)
plt.title('Color Image with Dimmed Square Region')
```

```python
# Now brighten a region in the color image, clipping at 255
color_image_copy2 = color_image.copy()
brighter_region = np.clip(color_image_copy2[40:100, 150:300] * 1.3, 0, 255)
color_image_copy2[40:100, 150:300] = brighter_region.astype(int)
plt.figure()
plt.imshow(color_image_copy2)
plt.title('Color Image with Brightened Square Region')
```
