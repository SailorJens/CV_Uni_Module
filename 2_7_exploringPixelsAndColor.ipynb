```python
import numpy as np
```

```python
# Create a 3x4 RGB image with combinations of red, green, blue
color_image = np.array([[[255, 0, 0], [0, 255, 0], [0, 0, 255], [255, 255, 0]], # Row 1: Red, Green, Blue, Yellow
                        [[0, 255, 255], [255, 0, 255], [128, 128, 128], [0, 0, 0]], # Row 2: Cyan, Magenta, Gray, Black
                        [[50, 100, 150], [200, 50, 0], [0, 150, 50], [250, 250, 250]]], # Row 3: Various Colors, White
                        dtype=np.uint8)
print("Color Image:\n", color_image)
```

```python
color_image2 = np.array([
    [[0, 0, 0], [0, 0, 255], [0, 255, 0], [0, 255, 255]], 
    [[255, 0, 0], [255, 0, 255], [255, 255, 0], [255, 255, 255]], 

    ], 
    dtype=np.uint8
    )
```

```python
# display the color image
import matplotlib.pyplot as plt
plt.imshow(color_image2)
plt.title('3x4 RGB Color Image')
plt.axis('off')
plt.show()
```

```python
def to_base3_3digits(n):
    digits = [0, 0, 0]
    for i in range(2, -1, -1):
        digits[i] = n % 3
        n //= 3
    return digits

image = np.zeros((9, 3, 3), dtype=np.uint8)

for i in range(9):  # 0 to 26
    for j in range(3):
        for k in range(3):
            image[i, j, k] = int((to_base3_3digits(i*3 + j)[k] * 255 / 2))
print("Generated 9x3 RGB Image with Base-3 Color Combinations:\n")

plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.title('9x3 RGB Image with Base-3 Color Combinations')
plt.axis('off')
plt.show()

print(f"Dimension: {image.ndim}")
```

```python
# creating a 3x3 binary image with True and False
binary_image = np.array([[True, False, True],
                         [False, True, False],
                         [True, True, False]], dtype=bool)
plt.imshow(binary_image, cmap='gray')
plt.title('3x3 Binary Image')
plt.axis('off')
plt.show()
```
