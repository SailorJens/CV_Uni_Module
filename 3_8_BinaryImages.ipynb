```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('PCB.jpg')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
grey_hist = cv2.calcHist([image_gray], [0], None, [256], [0, 256])
plt.plot(grey_hist)
plt.title('Grayscale Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.xlim([0, 256])
plt.grid(True, alpha=0.3)
```

```python
threshold_value = 200 # based on histogram analysis
ret, binary_image = cv2.threshold(image_gray, threshold_value, 255, cv2.THRESH_BINARY)
# ret is the threshold used, binary_image is the resulting binary image
# there are other methods like cv2.THRESH_BINARY_INV, cv2.THRESH_OTSU, etc.
# These also work:
#binary_image = (image_gray >= threshold_value) * 255
# This works too when displaying with plt or cv2, but produces a binary image with values 0 and 1, not 255
# so when saving it would be wrong
#binary_image = (image_gray >= threshold_value).astype(np.uint8)

plt.imshow(binary_image, cmap='gray')
```
