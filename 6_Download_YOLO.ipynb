## Step 2: Verify Installation and download models

```python
from ultralytics import YOLO

ultralytics.__version__
```

```python


# Detection model
model_det = YOLO("yolo26n.pt")

# Segmentation model
model_seg = YOLO("yolo26n-seg.pt")

print("Models loaded (and cached).")
```

```python
# Check MPS is available
import torch
torch.backends.mps.is_available()
```

## Step 3 Training a YOLO model

```python
# Train the model
results = model_det.train(
    data="african-wildlife.yaml", 
    epochs=100, 
    imgsz=640,
    device="mps"
    )
```

So the training took close to 4 hours for 100 epochs of 66 RGB jpg images.
